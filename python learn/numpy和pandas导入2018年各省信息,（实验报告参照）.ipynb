## 用numpy输入数据并计算人均GDP

```python
import numpy as np
data = np.loadtxt("data.csv",encoding='utf-8',delimiter=",",dtype=object)
print(data)
GDP = np.array(data[1:,2],dtype="float")
population = np.array(data[1:,1],dtype="float")
print("\n\n各省份人均GDP:\n",np.c_[data[1:,0],GDP/population])
```
