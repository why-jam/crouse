# 多类型的画图

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(10,8),dpi = 100)

plt.subplot(2,2,1)
x = np.linspace(0,10,100)
axes = fig.gca()
axes.set_xlabel('x')

axes.set_title('sin and cos')
plt.plot(x,np.cos(x),'r',linestyle='-.',linewidth =1.0,label = 'cos')
plt.plot(x,np.sin(x),'g',linestyle='-',linewidth =1.0,label = 'sin')
plt.legend()

plt.subplot(2,2,2)
rng = np.random.RandomState(0)
a = rng.randn(100)
b = rng.randn(100)
colors = rng.rand(100)
sizes = 100*rng.rand(100)
plt.scatter(a,b,c = colors,s = sizes,alpha = 0.3)

plt.subplot(2,2,3)
a = np.random.rand(100)
b = np.random.rand(100)
plt.scatter(a,b)

plt.subplot(2,2,4)
plt.plot(x,np.exp(x),'k',linestyle='-.',linewidth =1.0,label = 'e(x)')
plt.legend()

plt.show()
```

# 饼状图

```python
plt.figure(figsize=(16,12))
with open('data2.txt') as f:
    lines=f.readlines()

sizes=np.empty((18,9))
for i in range(18):
    sizes[i]=np.array(lines[i].split(','),dtype=np.float)
    plt.subplot(3,6,i+1)
    colors=['red','yellow','c','violet','r','grey','lime','cyan','y']
    explode=(0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)
    plt.pie(sizes[i],explode=explode,labels = ['sleep','2','3','4','5','6','7','8','9'],colors=colors,labeldistance=1.1,autopct='%3.1f%%',startangle=90,pctdistance=0.6)
#for t in l_text:
#    t.set_size=(30)
#for t in p_text:
#    t.set_size=(20)
plt.axis('equal')
plt.legend()
plt.show()
```

# csv文件中的导入和饼状图

```python
import numpy as np
data = np.loadtxt("data.csv",encoding='utf-8',delimiter=",",dtype=object)

plt.figure(figsize=(6,4),dpi=60)
a = np.empty((4,4))
for i in range(1,5):
    a[i-1]=np.array(data[i][1:],dtype=np.float)
    print(a[i-1])

explode=(0,0,0,0.1)
colors=['lime','cyan','pink','green']
labels=['sleep','entertainment','lesson','walk']
plt.pie(a[1],explode=explode,colors=colors,labels=labels,startangle=90,autopct='%3.2f%%',pctdistance=0.6,labeldistance=1.1)
plt.axis('equal')
plt.legend(loc='lower right')
```
