# 多层索引表示省2018年到2010年的GDP和人口

```python
import pandas as pd
import numpy as np
index = pd.MultiIndex.from_product([['北京市','天津市','河北省','山西省','内蒙古自治区','辽宁省','吉林省','黑龙江省','上海市'], ['GDP','人口']],
                                   names=['省', '数据'])
columns = pd.MultiIndex.from_product([[2018,2017,2016,2015,2014,2013,2012,2011,2010]],names=['年份'])
```

### 从文件中读取数据

```python
with open('data2.txt') as f:  
    lines=f.readlines()
    data0=np.zeros((18,9))
    for i in range(len(lines)):
        data0[i]=np.array(lines[i].split(','),dtype=np.float)
```

```python
pop= pd.DataFrame(data0,index=index,columns=columns)
pop
```

```python
pop.unstack()
```
