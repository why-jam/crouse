### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
import requests
import re
#根据包含的正则表达式的字符串创建模式对象
r = re.compile(r'<a.*href=\"(https?://.*?)[\"|\'].*')
#禁用警告
import urllib3
urllib3.disable_warnings()

def geturlsdeep(seed,depth):
    #根据包含的正则表达式的字符串创建模式对象
    r = re.compile(r'href=[\'"]?(http[^\'">]+)')
    #深度字典，{url:层级，…},用来去重
    stack = [seed]
    storage = {}
    storage[seed] = 0
    used=set()
    while len(stack)>0 :
        try:
            url = stack.pop(-1)
            #获取种子网页源代码
            html = requests.get(url,timeout=5,verify=False).text
            print(storage[url],url)
            #找到所有与self.__url匹配的字符串
            new_urls = re.findall(r,html)
            for new_url in new_urls:
                #判断深度字典是否超出深度
                if storage[url]<depth:
                    storage[new_url] = storage[url]+1
                    stack.append(new_url)      
        except Exception as e:
            pass
    
if __name__ == '__main__':
    seed = 'http://httpbin.org'
    #一共爬取2级
    geturlsdeep(seed,depth=2)
```

```python
深度遍历，先找到一级然后继续往下寻找，直到到达指定深度，返回上一层继续深度遍历
```

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

### 去重

```python
import requests
import re
#根据包含的正则表达式的字符串创建模式对象
r = re.compile(r'<a.*href=\"(https?://.*?)[\"|\'].*')
#禁用警告
import urllib3
urllib3.disable_warnings()

#临时存放url的队列
tmplist = []
#存放爬取到的所有url
urlList= []
#深度字典
depthDict = {}

def geturls(depth):
    while(len(tmplist)>0):
        #从临时列表中移除并获取对首元素链接
        url = tmplist.pop(0)
        #添加到链接列表
        urlList.append(url)
        if depthDict[url] < depth:
            #获取子列表url列表
            html = requests.get(url).text
            sublist = re.findall(r,html)
            for u in sublist:
                #判断深度字典有无此键，达到去重目的
                if u not in depthDict :
                    depthDict[u] = depthDict[url]+1
                    tmplist.append(u)

if __name__ == '__main__':
    #设置种子网站以及层次
    seed = 'http://httpbin.org'
    depthDict[seed] = 0
    tmplist.append(seed)
    geturls(2)
    for url in urlList:
         print(depthDict[url],url)
```

### 不去重

```python
import requests
import re
#根据包含的正则表达式的字符串创建模式对象
r = re.compile(r'<a.*href=\"(https?://.*?)[\"|\'].*')

import urllib3
urllib3.disable_warnings()

#临时存放url的队列
tmplist = []
#存放爬取到的所有url
urlList= []
#深度字典
depthDict = {}

def geturls(url,depth):
    depthDict[url]=0
    tmplist.append(url)
    while(len(tmplist)>0):
        #从临时列表中移除并获取对首元素链接
        url = tmplist.pop(0)
        #添加到链接列表
        urlList.append(url)
        if depthDict[url] < depth:
            #获取子列表url列表
            html = requests.get(url).text
            sublist = re.findall(r,html)
            #遍历全部子列表
            for u in sublist:
                depthDict[u] = depthDict[url]+1
                tmplist.append(u)

if __name__ == '__main__':
    seed = 'http://httpbin.org'
    geturls(seed,depth = 2)
    for url in urlList:
         print(depthDict[url],url)
```

### (选)将抓取网页的去重深度遍历算法封装成对象（类），并测试。 

```python
class Rela:
    def __init__(self,url,depth):
        self.__url = url
        self.__depth = depth
    def geturlsdeep(self):
        import requests
        import re
        count = self.__depth
        #根据包含的正则表达式的字符串创建模式对象
        r = re.compile(r'<a.*href=\"(https?://.*?)[\"|\'].*')
        seed = self.__url
        stack = [seed]
        #存放抓取的网页
        storage = {}
        storage[seed] = 0
        #集合用于去重
        used=set()
        while len(stack)>0 :
            try:
                #输出栈的最后一个元素
                url = stack.pop(-1)
                #获取种子网页源代码
                html = requests.get(url,timeout=5,verify=False).text
                print("\t"*storage[url],"%d:%s"%(storage[url],url))
                used.add(url)
                #找到所有与self.__url匹配的字符串
                new_urls = re.findall(r,html)
                for new_url in new_urls:
                    #判断深度字典有无此键，达到去重目的并控制抓取深度
                    if new_url not in used and new_url not in stack and storage[url]<count:
                        storage[new_url] = storage[url]+1
                        stack.append(new_url)      
            except Exception as e:
                pass

#禁用警告        
import urllib3
urllib3.disable_warnings()
```
